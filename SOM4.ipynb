```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Credit_Card_Applications.csv')
```

```python
dataset
```

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
X= sc.fit_transform(X)
```

```python
X
```

```python
from minisom import MiniSom
```

```python
som=MiniSom(x=10,y=10,input_len=15,sigma=1.0,learning_rate=0.5)
```

```python
som.random_weights_init(X)
```

```python
som.train_random(data=X,num_iteration=100)
```

```python
from pylab import bone,pcolor,colorbar,plot,show
```

```python
bone() # used to initilize a blank/white window which will contain the map
pcolor(som.distance_map().T) # distance_map Method will return all the Mean Interneuron Distances in one matrix. So this will return the matrix of all these distances for all the winning nodes.(.T is used for transpose method)
colorbar()
markers=['o','s'] #here o means circle and s means square
colors = ['r', 'g']
for i, x in enumerate(X):
    w = som.winner(x)
    plot(w[0] + 0.5,
         w[1] + 0.5,
         markers[y[i]],
         markeredgecolor = colors[y[i]], # here we used the y (dependent variable)
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2)
show()
```

```python
mappings=som.win_map(X)
```
